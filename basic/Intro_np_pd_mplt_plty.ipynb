## Introduction à Pandas

```python
import pandas as pd

data = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(data)
```

```python
data2 = {
    'Nom': ['Alice', 'Bob', 'Charlie'],
    'Âge': [25, 30, 35],
    'Ville': ['Paris', 'Lyon', 'Marseille']
}

df = pd.DataFrame(data2)
print(df)
```

```python
df.groupby('Ville')['Âge'].mean()   # Moyenne d'âge par ville
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Nom': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 3], 'Salaire': [3000, 4000, 5000]})

merged_df = pd.merge(df1, df2, on='ID', how='inner')   # Jointure sur 'ID'
print(merged_df)

```

## Introduction à Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
x = np.linspace(0, 10, 100)  # Génération de 100 points entre 0 et 10
y = np.sin(x)  # Calcul de sin(x) pour chaque valeur de x

x, y
```

```python

plt.plot(x, y, label='sin(x)')  # Tracé de la courbe avec un label
plt.xlabel("x")  # Ajout d'un label pour l'axe des x
plt.ylabel("y")  # Ajout d'un label pour l'axe des y
plt.title("Exemple de courbe avec Matplotlib")  # Titre du graphique
plt.legend()  # Affichage de la légende
plt.show()  # Affichage du graphique
```

```python
plt.plot(x, y, linestyle='--', color='r', marker='o', label='sin(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x, y)
plt.text(8, 0, "X", fontsize=12, color='red')  # Annotation de X au point de coordonnées (8, 0)
plt.show()
```

```python
data = np.random.randn(1000)  # Génération de 1000 nombres aléatoires
plt.hist(data, bins=30, color='red', alpha=0.7)
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, color='green')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)
x = np.linspace(0, 10, 100)

defs = [np.sin, np.cos, np.tan, np.exp]
titles = ["sin(x)", "cos(x)", "tan(x)", "exp(x)"]

for ax, func, title in zip(axs.ravel(), defs, titles):
    ax.plot(x, func(x))
    ax.set_title(title)

plt.tight_layout()
plt.show()
```
