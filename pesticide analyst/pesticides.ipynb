# Analyse du fichier pesticides.csv

### Objectif: analyser l'utilisation des pesticides dans différents pays, en fonction de l'année

Etapes:

Charger le fichier

Aperçu des données

Analyser l'évolution de l'utilisation des pesticides dans le temps pour un pays donné (Filtrer pour un pays spécifique (exemple: Albania))

Visualiser l'évolution de l'utilisation des pesticides


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Charger le fichier


df = pd.read_csv('../pesticide/pesticides.csv')
df
```

```python
# Apercu des données

df = df.head()
df
```

```python

#Analyser l'évolution de l'utilisation des pesticides dans le temps 
# pour un pays donné (Filtrer pour un pays spécifique (exemple: Albania))

albania_pesticides = df[df['Area'] == 'Albania']
albania_pesticides.head()
```

```python
#Visualiser l'évolution de l'utilisation des pesticides
plt.plot(albania_pesticides['Year'],
         albania_pesticides['Value'])

plt.title('Analyse des quantité de pesticides utilisées en Albanie')
plt.xlabel('Année')
plt.ylabel('Tonnes de pesticides')

plt.show()
```
