```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Chargement et exploration des données 

rainfall_df = pd.read_csv('../pesticide/rainfall.csv')
```

```python
print(rainfall_df.head())
```

```python
#Analyser les précipitations moyennes 
# pour un pays donné (Filtrer pour un pays spécifique (exemple: Afghanistan))
afghanistan_rainfall = rainfall_df[rainfall_df['Area'] == 'Afghanistan']
afghanistan_rainfall
```

```python
#Visualisation des précipitations au fil des ans
plt.plot(afghanistan_rainfall['Year'],
        afghanistan_rainfall['average_rain_fall_mm_per_year'])
plt.title('Précipitations moyennes en Afghanistan')
plt.xlabel('Année')
plt.ylabel('Précipitations (mm)')
plt.show()
```
