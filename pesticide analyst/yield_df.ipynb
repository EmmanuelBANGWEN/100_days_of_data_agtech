# Objectif: Analysez le rendement des cultures et les facteurs qui peuvent l'influencer, comme les précipitations et l'utilisation des pesticides


Etapes:

Chargement des données 

Aperçu des données

Analyser l'impact des précipitations et des pesticides sur le rendement (exemple pour le maïs en Albanie)

Visualisation des rendements par rapport aux précipitations et pesticides

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Chargement des données 
df_yield_df = pd.read_csv('../pesticide/yield_df.csv')
```

```python

#Aperçu des données
df_yield_df.head()
```

```python
df_yield_df.columns
```

```python

#Analyser l'impact des précipitations et des pesticides sur le rendement 
# (exemple pour le maïs en Albanie)
albania_mais =df_yield_df[(df_yield_df['Area']=='Albania') & (df_yield_df['Item']=='Maize')]
albania_mais
```

```python

#Visualisation des rendements par rapport aux précipitations et pesticides
fig, ax1 = plt.subplots()

ax1.set_xlabel('Année')
ax1.set_ylabel('Rendement (hg/ha)')
ax1.plot(albania_mais['Year'], albania_mais['hg/ha_yield'], color='tab:blue', label="Rendement")

ax2 = ax1.twinx()
ax2.set_ylabel('Pesticides (tonnes)', color='tab:green')
ax2.plot(albania_mais['Year'], albania_mais['pesticides_tonnes'], color='tab:green', label="Pesticides")

ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # décaler la troisième échelle
ax3.set_ylabel('Temperature (celcius)', color='tab:red')
ax3.plot(albania_mais['Year'], albania_mais['avg_temp'], 
         color='tab:red', label="Temperature")
plt.title('Rendement du Maïs en Albanie')


plt.show
```
