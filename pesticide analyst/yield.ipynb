```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement des données 
df_yield = pd.read_csv('../pesticide/yield.csv')
```

```python
# Analyser l'impact des précipitations et des pesticides sur le rendement (exemple pour le maïs en Albanie)
maid_yield = df_yield[(df_yield['Area'] == 'Afghanistan') & (df_yield['Item'] == 'Maize')]
maid_yield.head()
```

```python
# Visualisation des rendements par rapport aux précipitations et pesticides
plt.plot(maid_yield['Year'], maid_yield['Value'])
plt.title('Rendement du Maïs en Afghanistan')
plt.xlabel('Année')
plt.ylabel('Rendement (hg/ha)')
plt.show()
```
