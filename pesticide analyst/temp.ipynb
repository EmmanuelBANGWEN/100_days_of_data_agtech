Analyser la température moyenne pour un pays spécifique (exemple: Côte d'Ivoire)

Visualisation de la température moyenne


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement des données
df_temp = pd.read_csv('../pesticide/temp.csv')
```

```python
# Aperçu des données
df_temp
```

```python
# Analyser la température moyenne pour un pays spécifique (exemple: Côte d'Ivoire)
df_temp_ivory_coast = df_temp[df_temp['country']=="Côte D'Ivoire"]
df_temp_ivory_coast
```

```python
new_df = df_temp_ivory_coast.fillna(0)
new_df
```

```python
# Visualisation de la température moyenne
plt.plot(new_df['year'],
        new_df['avg_temp'])
plt.title('Visualisation de la température moyenneen cote divoire')
plt.xlabel('Année')
plt.ylabel('Temperature')
```
